DML:
DML stands for Data Manipulation Language. It is a subset of SQL (Structured Query Language) used for modifying and manipulating data within a database.

INSERT: The INSERT statement is used to insert new rows of data into a table.
INSERT INTO users (name, age) VALUES ('John Doe', 25);

UPDATE:
The UPDATE statement is used to modify existing data in a table
Example:
PDATE users SET age = 26 WHERE name = 'John Doe';

DELETE:
The DELETE statement is used to remove rows from a table that meet certain conditions.
Example:
DELETE FROM users WHERE age > 30;

DQL:
DQL (Data Query Language) is a term sometimes used to refer to the subset of SQL (Structured Query Language) that is specifically focused on querying and retrieving data from a database.


SELECT:
In SQL, the SELECT statement is the primary DQL statement used for querying data.


Example:
SELECT name, age, city FROM users WHERE age > 25;

Primary Key:
A primary key is a unique identifier for a record or row in a database table. It is a column or a set of columns that uniquely identifies each row in the table. The primary key enforces data integrity by ensuring that each record is uniquely identifiable.

Foreign Key:



A foreign key is a column or a set of columns in a database table that  a relationship between two tables. It creates a logical connection between the child table and parent table. The foreign key in the child table refers to the primary key of the parent table, thereby defining a relationship between them.
